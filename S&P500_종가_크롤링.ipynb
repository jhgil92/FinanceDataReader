```
import FinanceDataReader as fdr
```

```
# S&P500 종목 리스트
sp500 = fdr.StockListing('S&P500')
sp500.head(10)
```

```
# S&P500 종목은 실제 505개
len(sp500)
```

```
import pandas as pd
df_price = pd.DataFrame()  

for ticker, name in zip(sp500.Symbol, sp500.Name):
  tmp = fdr.DataReader(ticker, "2019-08-30", "2019-08-30")
  tmp['Symbol'] = ticker
  tmp['Name'] = name
  df_price = pd.concat([df_price, tmp])
  print(ticker, "done")
```
